```python
# Load and process the expiry data with updated columns
df_expiry = pd.read_csv('expiry.csv')

# Convert the relevant columns to datetime format
df_expiry['FTD'] = pd.to_datetime(df_expiry['FTD'], errors='coerce')  # First Trading Day
df_expiry['LTD'] = pd.to_datetime(df_expiry['LTD'], errors='coerce')  # Last Trading Day

# Load the main cocoa futures data (assuming it contains Date and CLOSE columns)
# Here I'll create a sample DataFrame to illustrate, as we only have one CSV file provided
df_main = pd.DataFrame({
    "Date": pd.date_range(start="2023-01-01", periods=100, freq="D"),
    "CLOSE": pd.Series([2500 + i for i in range(100)])
})
df_main.set_index("Date", inplace=True)

# Create the line graph for 'CLOSE' prices
fig = go.Figure()

# Plot the cocoa futures closing prices
fig.add_trace(go.Scatter(
    x=df_main.index,
    y=df_main['CLOSE'],
    mode='lines',
    name='Cocoa Futures Close Price'
))

# Add shaded regions for each contract period
for i in range(len(df_expiry)):
    contract_start = df_expiry['FTD'].iloc[i]
    contract_end = df_expiry['LTD'].iloc[i]
    if pd.notna(contract_start) and pd.notna(contract_end):
        fig.add_trace(go.Scatter(
            x=[contract_start, contract_end, contract_end, contract_start],
            y=[df_main['CLOSE'].min(), df_main['CLOSE'].min(), df_main['CLOSE'].max(), df_main['CLOSE'].max()],
            fill='toself',
            fillcolor='rgba(0, 100, 200, 0.1)',
            line=dict(color='rgba(0,0,0,0)'),
            name=f'Contract {i+1}'
        ))

# Customize layout
fig.update_layout(
    title="Cocoa Futures Closing Prices with Contract Expiry Dates",
    xaxis_title="Date",
    yaxis_title="Closing Price",
    template="plotly_white"
)

# Show plot
fig.show()
```
